#  2.9: Plotly — Avanzado

Gráficos 3D, mapas, treemaps y exportación.

```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd
```

## 6.1 Superficie 3D

```python
x = np.linspace(-5, 5, 60)
y = np.linspace(-5, 5, 60)
X, Y = np.meshgrid(x, y)
# np.meshgrid(x, y) → crea cuadrícula 2D a partir de arrays 1D
# X: cada fila es copia de x
# Y: cada columna es copia de y
# Necesario para evaluar funciones en un plano 2D

Z = np.sin(np.sqrt(X**2 + Y**2))
# Función: sin(r) donde r = distancia al origen
# Crea el patrón de ondas en el agua

fig = go.Figure(data=[go.Surface(
    x=x, y=y, z=Z,
    colorscale='Viridis',
    opacity=0.9,
    contours=dict(
        x=dict(show=True, color='white', width=1),
        y=dict(show=True, color='white', width=1),
        z=dict(show=True, color='white', width=1, usecolormap=True)))])

fig.update_layout(
    title='Superficie 3D: f(x,y) = sin(sqrt(x^2+y^2))',
    scene=dict(
        # 'scene' controla el espacio 3D
        xaxis_title='X', yaxis_title='Y', zaxis_title='Z',
        camera=dict(eye=dict(x=1.5, y=1.5, z=1.0))),
    # camera.eye → posición de la cámara (x,y,z)
    template='plotly_dark',
    width=800, height=550)

fig.show()
# Arrastra el mouse para rotar la superficie
```

## 6.2 Scatter 3D

```python
df_iris = px.data.iris()

fig = px.scatter_3d(
    df_iris,
    x='sepal_length', y='sepal_width', z='petal_length',
    color='species',
    symbol='species',    # símbolo diferente por especie
    size='petal_width',
    size_max=12, opacity=0.8,
    title='Dataset Iris en 3D',
    template='plotly_white')

fig.update_layout(scene=dict(
    xaxis_title='Longitud Sepalo (cm)',
    yaxis_title='Ancho Sepalo (cm)',
    zaxis_title='Longitud Petalo (cm)'))

fig.show()
```

## 6.3 Treemap y Sunburst

```python
df_gap = px.data.gapminder()
df_2007 = df_gap[df_gap['year'] == 2007].copy()

# Treemap — rectangulos anidados (ideal para jerarquías y proporciones)
fig_tree = px.treemap(
    df_2007,
    path=[px.Constant('Mundo'), 'continent', 'country'],
    # path → jerarquía de niveles del treemap
    # px.Constant → nodo raíz con nombre fijo
    values='pop',        # tamaño de cada rectángulo
    color='lifeExp',     # color según esta variable
    color_continuous_scale='RdYlGn',  # colormap
    title='Población Mundial por País (2007)')
fig_tree.show()

# Sunburst — como treemap pero circular
fig_sun = px.sunburst(
    df_2007,
    path=['continent', 'country'],
    values='pop', color='lifeExp',
    color_continuous_scale='RdYlGn',
    title='Sunburst: Poblacion y Esperanza de Vida')
fig_sun.show()
```

## 6.4 Mapa coroplético

```python
df_gap = px.data.gapminder()
df_2007 = df_gap[df_gap['year'] == 2007]

fig = px.choropleth(
    df_2007,
    locations='iso_alpha',    # columna con códigos ISO de países (3 letras)
    color='lifeExp',          # variable para colorear los países
    hover_name='country',
    color_continuous_scale='Viridis',
    title='Esperanza de Vida por País (2007)',
    template='plotly_white')

fig.update_layout(
    geo=dict(
        showframe=False,      # ocultar marco del mapa
        showcoastlines=True,
        projection_type='natural earth'))
    # projection_type: 'equirectangular','mercator','orthographic','natural earth'

fig.show()
```

## 6.5 Exportar gráficos

```python
fig = px.scatter(px.data.iris(), x='sepal_length', y='sepal_width',
                 color='species', title='Iris Dataset')

# HTML interactivo — el más útil para compartir
fig.write_html(
    'grafico_interactivo.html',
    include_plotlyjs=True,  # True = incluye Plotly.js (portable)
    full_html=True,         # True = HTML completo | False = solo el div
    auto_open=False)

# JSON — para reutilizar la figura en otro código
fig.write_json('grafico.json')

# PNG/PDF (requiere: pip install kaleido)
# fig.write_image('grafico.png', width=1200, height=700, scale=2)
# scale=2 → doble resolución (pantallas de alta densidad)

# Cargar desde JSON
import plotly.io as pio
# fig_cargada = pio.read_json('grafico.json')

print('Exportado: grafico_interactivo.html y grafico.json')
```
